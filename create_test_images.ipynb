```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#set video file
video_name = "harder_challenge_video.mp4"

#Open the video file
cap = cv2.VideoCapture(video_name)
print(type(cap))
```

```python
n_time = 41.6 # time of the frame in seconds
show_time  = n_time * 1000 # set time in miliseconds

#The first argument of cap.set(), number 0 defines that parameter for setting the frame selection in miliseconds.
# number 0 means CV_CAP_PROP_POS_MSEC Current position of the video file in milliseconds.
#The second argument defines the frame number in miliseconds
cap.set(0,show_time);

#Read the next frame from the video. If you set frame 749 above then the code will return the last frame.
ret, frame = cap.read()
# colors are wrong because of the another colos-space as we read img by cv2 not matplot but it does not matter
plt.figure()
plt.imshow(frame)
```

```python
cv2.imwrite("harder_challenge_test3.jpg", frame)
```
